# Demo: Cylinder Skeletonization


```python
import sys, os
sys.path.append(os.path.abspath('..'))

from mcf2swc import MeshManager, PolylinesSkeleton, fit_swc, SWCFitOptions

import logging
logging.basicConfig(level=logging.INFO)
```

## Create and Visualize Cylinder

```python
# Load cylinder mesh we created in demo_cylinder_mesh.ipynb

mm = MeshManager()
mm.load_mesh("../data/demo/cylinder.obj")
mm.print_mesh_analysis()
mm.visualize_mesh_3d(title="Original Cylinder", backend="plotly")
```

## Skeletonize the Mesh

```python
pls = PolylinesSkeleton().from_txt("../data/demo/cylinder.polylines.txt")
skel = fit_swc(mm, pls, options=SWCFitOptions(spacing=1))
skel.to_swc_file("../data/demo/cylinder.swc")
skel.print_attributes()
print(skel)
```

# Plot cross sections and fitting disks

```python
# plot using swctools
from swctools import SWCModel, plot_model

model = SWCModel.from_swc_file("../data/demo/cylinder.swc")
model.print_attributes()
plot_model(swc_model=model)
```

## Summary

The cylinder segmentation shows:
- **Simple topology**: Each slice typically contains one segment
- **Linear connectivity**: Segments connect in sequence along the z-axis
- **Volume distribution**: End caps may have smaller volumes due to geometry

This represents the baseline case for mesh segmentation with genus = 0 topology.
