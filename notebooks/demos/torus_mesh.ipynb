# Demo: Torus Mesh

Create and visualize torus (donut-shaped) mesh geometries using mcf2swc.

Resulting mesh saved to `../data/demo/torus.obj` for skeletonization in CGAL.

## Setup

```python
import numpy as np
import trimesh
import plotly.graph_objects as go

# Import mcf2swc functions
from mcf2swc import MeshManager, example_mesh

import logging
logging.basicConfig(level=logging.INFO)

print("✅ Libraries imported successfully!")
```

## Create Torus

With a mesh loaded in a `MeshManager`, we can apply transformations to it. Here, we create a torus mesh and apply a rotation to it.

```python
# Create a torus mesh
torus = example_mesh("torus")

# Rotate torus 90 degrees, so it's standing on its side
torus.apply_transform(trimesh.transformations.rotation_matrix(
    angle=np.pi / 2,
    direction=[1, 0, 0],
    point=torus.centroid
))

print(f"Created torus with {len(torus.vertices)} vertices and {len(torus.faces)} faces")

mm = MeshManager(torus)
```

## 3D Visualization

```python
mm.visualize_mesh_3d()
```

## Interactive Cross-Sections

```python
mm.visualize_mesh_slice_interactive()
```

## Mesh Properties

```python
mm.print_mesh_analysis()
```

## Save mesh to .obj file

```python
savefile = "../data/demo/torus.obj"
mm.save(savefile)
```
