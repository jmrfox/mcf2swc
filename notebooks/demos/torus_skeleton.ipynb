# Demo: Torus Skeletonization


```python
import sys, os
sys.path.append(os.path.abspath('..'))

from mcf2swc import MeshManager, PolylinesSkeleton, fit_morphology, FitOptions

import logging
logging.basicConfig(level=logging.DEBUG)
```

# Create and Visualize Torus

```python
# Load cylinder mesh we created in demo_cylinder_mesh.ipynb

mm = MeshManager()
mm.load_mesh("../data/demo/torus.obj")
mm.print_mesh_analysis()
mm.visualize_mesh_3d(title="Original Torus", backend="plotly")
```

# Plot mesh with skeleton

```python
pls = PolylinesSkeleton().from_txt("../data/demo/torus.polylines.txt")
mm.visualize_mesh_3d(polylines=pls)
```

## Fit Morphology model to Mesh

```python
morph = fit_morphology(mm, pls, options=FitOptions(spacing=1))
morph.to_swc_file("../data/demo/torus.swc")
morph.print_attributes()
```

# Plot fitted SWC model

```python
# plot using swctools
from swctools import SWCModel, plot_model

model = SWCModel.from_swc_file("../data/demo/torus.swc")
plot_model(swc_model=model)
```
