# Demo: Torus Skeletonization


```python
import sys, os
sys.path.append(os.path.abspath('..'))

from mcf2swc import MeshManager, PolylinesSkeleton, build_traced_skeleton_graph, TraceOptions

import logging
logging.basicConfig(level=logging.INFO)
```

# Create and Visualize Torus

```python
# Load cylinder mesh we created in demo_cylinder_mesh.ipynb

mm = MeshManager()
mm.load_mesh("../data/demo/torus.obj")
mm.print_mesh_analysis()
mm.visualize_mesh_3d(title="Original Torus", backend="plotly")
```

## Skeletonize the Mesh

```python
pls = PolylinesSkeleton().from_txt("../data/demo/torus.polylines.txt")
skel = build_traced_skeleton_graph(mm, pls, options=TraceOptions(spacing=1))
skel.to_swc_file("../data/demo/torus.swc")
print(skel)
```

# Plot cross sections and fitting disks

```python
# plot using swctools
from swctools import SWCModel, plot_model

model = SWCModel.from_swc_file("../data/demo/torus.swc")
plot_model(swc_model=model)
```
